```python
#importing necessary libraries
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
covid_india = pd.read_csv('DataVisualization/covid_19_india.csv') #statewise tally of COVID-19 cases in India
indiaDaily = pd.read_csv('DataVisualization/dailycases.csv') #Daily cases in india from the starting date till end of lockdown-1 [14.04.2020]
```

```python
covid_india = covid_india.drop(['Time', 'Sno'], axis=1) #Time of data retrival and S.No are un-necessary
```

```python
covid_india = covid_india[0:959] #taking cases from the begining till the end of lockdown-1
```

```python
covid_india = covid_india.rename(columns={"State/UnionTerritory": "State"}) #renaming columns for simplicity
covid_india = covid_india.drop(['ConfirmedIndianNational', 'ConfirmedForeignNational'], axis=1) #droping column with number of cases based on nationality
```

```python
fig = plt.figure() #creating a figure so that graph comes in place
ax = plt.gca() #get current axis
indiaDaily.plot(kind='line',x='Date',y='DailyCases',ax=ax) #plotting a line graph
plt.show()
plt.xticks(rotation=45)
plt.ylabel("Number of People")
plt.xlabel("Dates")
fig.suptitle('COVID-19 Daily Cases in India durind Lockdown Phase-1 ', fontsize=20)
```

```python
Delhi = covid_india[covid_india['State'] == 'Delhi'] #retrving data of Delhi
Kerala = covid_india[covid_india['State'] == 'Kerala'] #retreiving data of Kerala
Maharashtra = covid_india[covid_india['State'] == 'Maharashtra'] #retreiving data of Karnataka
```

```python
fig = plt.figure()
ax = plt.gca()
Delhi.plot(kind='line',x='Date',y='Confirmed',ax=ax ,label = 'Delhi')
plt.show()
plt.xticks(rotation=45)
plt.ylabel("Number of People")
plt.xlabel("Dates")
Kerala.plot(kind='line',x='Date',y='Confirmed',ax=ax, label = 'Kerala')
plt.show()
plt.xticks(rotation=45)
plt.ylabel("Number of People")
plt.xlabel("Dates")
ax.legend()
Maharashtra.plot(kind='line',x='Date',y='Confirmed',ax=ax, label = 'Maharastra')
plt.show()
plt.xticks(rotation=45)
plt.ylabel("Number of People")
plt.xlabel("Dates")
ax.legend()

fig.suptitle('Maharastra vs Delhi vs Kerala in controling COVID-19', fontsize=20)
```

This visualisation of data from 25th March to 14th March 2020 helps us to understand how well Kerala was able to control the COVID-19 outbreak, whereas other major states like Maharastra and Union territory like Delhi failed to control the outbreak, We cannot hold the Government responsible for this, It was supposed to be a collective responsibility of citizens, state government and the central government.
